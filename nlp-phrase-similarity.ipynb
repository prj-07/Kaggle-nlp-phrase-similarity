```python
from difflib import SequenceMatcher

def similar(a,b):
    return SequenceMatcher(None,a,b).ratio()
```

```python
str1 = test_df['target']
str2  = test_df['anchor']
```

```python
str3 =[]
for i in range(len(test_df)):
    a = similar(str1[i],str2[i])
    str3.append(a)
    
```

```python
str3 
```

```python
test_df['score'] = str3
```

```python
test_df.head()
```

```python
for i in range(len(test_df)):
    if(test_df['score'][i]<=0.15):
        test_df['score'][i] = 0
    if(test_df['score'][i]>=0.15 and test_df['score'][i]<0.40):
        test_df['score'][i]= 0.25
    if(test_df['score'][i]>=0.40 and test_df['score'][i]<0.60):
        test_df['score'][i]= 0.5
    if(test_df['score'][i]>=0.60 and test_df['score'][i]<0.80):
        test_df['score'][i]= 0.75
    if(test_df['score'][i]>=0.80 and test_df['score'][i]<1):
        test_df['score'][i]= 1
            
                
```
